```python
table1_data = []
table1_headers = ['name', 'size_of_the_bottle', 'MRP_of_the_bottle', 'price_of_the_bottle', 'Product_url']
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36",
    "Accept-Encoding": "gzip, deflate",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9",
    "DNT": "1",
    "Connection": "close",
    "Upgrade-Insecure-Requests": "1"
}


for i in range(1, 15):
    url = f"https://www.1mg.com/categories/homeopathy-57?filter=true&pageNumber={i}"
    page = requests.get(url, headers=headers)
    sleep(0.05)
    # Create BeautifulSoup object to parse the HTML
    soup1 = BeautifulSoup(page.content, "html.parser")
    soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
    sleep(0.05)
    a = soup2.find_all('div', class_ = 'col-md-3 col-sm-4 col-xs-6 style__container___1TL2R')  
    sleep(0.05)
    for item in a:
        if item.find(class_='style__mg-recommended___WgWac') is not None:
            continue  
        name = item.find(class_='style__pro-title___2QwJy').text.strip()

        pack_size = item.find(class_="style__pack-size___2JQG7").text.strip()

        mrp_element = item.find(class_='style__discount-price___25Bya')
        if mrp_element is not None:
            mrp = mrp_element.text.strip().replace('₹', '').strip()

        price = item.find(class_="style__price-tag___cOxYc").text.strip().replace('₹', '').strip()
        urlx = item.find('a')['href']
        product_url = 'https://www.1mg.com' + urlx
        table1_data.append([name, pack_size, mrp, price, product_url])
        sleep(0.05)  #
    print(f'page {i} done')
print("all Done")
```

```python
table1_df = pd.DataFrame(table1_data, columns=table1_headers)
```

```python
table2_data = []
table2_headers = ['name', 'brand_name','rating', 'number_of_rating']

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36",
    "Accept-Encoding": "gzip, deflate",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9",
    "DNT": "1",
    "Connection": "close",
    "Upgrade-Insecure-Requests": "1"
}

for i in range(1, 15):
    url = f"https://www.1mg.com/categories/homeopathy-57?filter=true&pageNumber={i}"
    page = requests.get(url, headers=headers)
    sleep(0.05)
    # Create BeautifulSoup object to parse the HTML
    soup1 = BeautifulSoup(page.content, "html.parser")
    soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
    sleep(0.05)
    product_containers = soup2.find_all('div', class_='col-md-3 col-sm-4 col-xs-6 style__container___1TL2R')
    sleep(0.05)
    for container in product_containers:
        if container.find(class_='style__mg-recommended___WgWac') is not None:
            continue
       
        urlx = container.find('a')['href']
        
        product_url = 'https://www.1mg.com' + urlx
        
        product_page = requests.get(product_url, headers=headers)
        product_soup = BeautifulSoup(product_page.content, "html.parser")
        
        
        name = product_soup.find(class_='ProductTitle__product-title___3QMYH').text.strip()
        
        brand_name = product_soup.find(class_='ProductTitle__manufacturer___sTfon').text.strip()
        
        rating = product_soup.find(class_='RatingDisplay__ratings-container___3oUuo').text.strip()
        
        number_of_ratings = product_soup.find(class_='RatingDisplay__ratings-header___ZNj5b').text.strip().replace(' Ratings','').strip()
        
        table2_data.append([name, brand_name, rating, number_of_ratings])
        
        sleep(0.05)
               
    print(f'page {i} done')
print("all Done")
```

```python
table2_df = pd.DataFrame(table2_data, columns=table2_headers)
```
